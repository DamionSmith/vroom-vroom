Univariate Analysis

```python

def count_column(df,col_list):
    count = {}
    for elem in col_list:
        count[elem] = df[elem].value_counts()
    return count


#We will drop 'latest_price' for the reasons above
columns_to_count = ['fuel-system', 'engine-type','engine-location','drive-wheels', 'body-style', 'make' ]
auto_string_counts = count_column(df, columns_to_count)
```

```python
plt.figure(figsize = (15,40))
plt.subplots_adjust(hspace = 1.5)

    
for n, key in enumerate(auto_string_counts):
    ax = plt.subplot(10,2,n+1)
    auto_string_counts[key].plot(kind = 'bar', ax = ax, title = key)
    
for tick in ax.xaxis.get_major_ticks()[1::2]:
        tick.set_pad(15);
```

```python
#Lets plot out the distribution of our prices and take a look
price_median = round(df['price'].median(),2)
fig = plt.figure(figsize=(12,12))
df['price'].plot(kind = 'density')
plt.axvline(price_median, c = 'r')
plt.annotate('Median: $'+str(price_median),
             xy = (price_median, plt.ylim()[1]*0.75),
             xytext = (price_median+ plt.xlim()[1]*0.1, plt.ylim()[1] * 0.8),
             textcoords = 'data',
             xycoords = 'data',
             arrowprops = dict(arrowstyle = '->', lw = 3, color = 'r'),
             fontweight = 'bold', fontsize = 'large')


plt.xlim(0,6500)
plt.xticks(np.arange(0,50000,3000));
plt.xlabel('Latest Price in $');
```

```python
#average price by brand
#Lets look at the average price by brand first and see what we get
auto_make_prices = df[['make','price']]
auto_average_make = auto_make_prices.groupby(by = 'make').agg('mean')
auto_average_make.sort_values(by = 'price', ascending = False)
```

```python
sns.set(font_scale = 1)
plt.subplots(figsize=(12,12))
sns.boxplot(x='make', y='price', data = auto_make_prices)
plt.xticks(rotation = 'vertical')
plt.ylabel('Price($)')
plt.xlabel('Make');
```
